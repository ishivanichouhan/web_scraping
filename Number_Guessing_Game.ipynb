```python
import random #random module use for generate random numbers
computer_number = random.randrange(1,101) #random.randrange  are use for range of  number for computer
user_input_number = int(input("Enter your number:")) # Take the number to the user
if user_input_number>computer_number: # if use for which number is high, user number or compuert number
  print("computer number is", computer_number)
  print("Your guess number is too high")
elif computer_number>user_input_number: #elif use for, if the previous conditions were not true, then try elif condition"
  print("computer number is", computer_number)
  print("Your guess number is too low")
else:                              # else condition use for if, if condition and elif condition are false so when use the  else condition
  print("computer number is", computer_number) #after this call the function and show the result
  print("Your guess number is equal")
```
